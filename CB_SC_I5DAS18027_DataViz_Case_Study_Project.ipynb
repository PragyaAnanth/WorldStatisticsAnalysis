```
df1=df.loc[(df["Country Name"]=="Brazil") | (df["Country Name"]=="India")]
df1
```

```
df1["Country Name2"]=df1["Country Name"].map({"India":1,"Brazil":2})
```

```
df1
```

```
dims=["Agriculture, forestry, and fishing, value added (% of GDP)",
      "Agricultural raw materials exports (% of merchandise exports)",
    'Agricultural raw materials imports (% of merchandise imports)']
```

```
fig=px.parallel_coordinates(df1,color="Country Name2",labels={"Country Name2":"Country"},dimensions=dims,
                            color_continuous_scale=px.colors.diverging.Tealrose)
                             #color_continuous_midpoint=2)
fig.show(renderer='colab')
```

```
df11=df.loc[df["Country Name"]=="India"]
```

```
dims1=[  'Access to electricity (% of population)',
       'Access to electricity, rural (% of rural population)',
       'Agricultural machinery, tractors']
```

```
fig1 = px.scatter_matrix(df11,dimensions=dims1,color="year")
fig1.update_yaxes(automargin=True,tickangle=45)
fig1.update_xaxes(automargin=True,tickangle=45)
fig1.show(renderer='colab')
```
